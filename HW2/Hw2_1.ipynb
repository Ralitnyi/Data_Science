```python
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
tables = pd.read_html(url, attrs={"class":"wikitable collapsible collapsed"}, match="Регіон", decimal=',')
table = tables[0]
table
```

Вивести перші рядки таблиці за допомогою методу head

Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
print('Head:\n', table.head())
print('Shape:\n', table.shape)
```

Замініть у таблиці значення "—" на значення NaN


```python
table.replace(to_replace="—", value=np.nan, inplace=True)
table
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes


```python
table.dtypes
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"


```python
table = table.astype({'2014': float, '2019': float})
table.dtypes
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)


```python
table.isnull().sum() / len(table) * 100
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці


```python
table.drop(27, inplace=True)
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)


```python
table_filled = table.fillna(table.loc[:, '1950':].mean())
table_filled
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні


```python
reigon_table = table_filled[table_filled['2019'].mean() < table_filled['2019']]['Регіон']
reigon_table
```

У якому регіоні була найвища народжуваність у 2014 році?


```python
table[table['2014'].max() == table['2014']]
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році


```python
import matplotlib.pyplot as plt
```

```python
table_2019 = table[['Регіон', '2019']]
plt.figure(figsize=(8, 4))
plt.bar(table_2019['Регіон'], table_2019['2019'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(fontsize=10, rotation=90)
plt.show()
```

```python
donetsk_values = table[table['Регіон'] == 'Донецька'].iloc[0, 1:].tolist()
years = table.columns[1:]
plt.style.use('ggplot')
plt.plot(years, donetsk_values, label='Birth rate in Donetsk', marker='D', color='red')
plt.title('Народжуваність в Донецьку')
plt.xticks(fontsize=12, color='red')
plt.yticks(fontsize=12, color='red')
plt.grid(color='black')
legend = plt.legend()
legend.get_texts()[0].set_color('red')
plt.show()
```

```python
kyiv_values = table[table['Регіон'] == 'Київська'].iloc[0, 1:].tolist()
plt.style.use('grayscale')
plt.plot(years, kyiv_values, marker='o', label='Birth rate in Kyiv region')
plt.title('Kyiv region')
plt.xlabel("Дата", fontsize="medium", color="black")
plt.ylabel("Температура", fontsize="medium", color="black")
plt.legend()
plt.show()
```
